```python
from matplotlib_venn import venn3, venn3_circles
from matplotlib import pyplot as plt
```

```python
def intersection_2(a, b):
    return a & b

def intersection_3(a, b, c):
    return a & b & c

def difference_2(a, b):
    return a.difference(b)

def difference_3(a, b, c):
    a_minus_b = a.difference(b)
    a_minus_c = a.difference(c)
    return intersection_2(a_minus_b, a_minus_c)
```

```python
def get_only_oo(oo, po, co):
    return len(difference_3(oo, po, co))

def get_only_po(po, co, oo):
    return len(difference_3(po, co, oo))

def get_only_co(co, oo, po):
    return len(difference_3(co, oo, po))

def get_oo_and_po(oo, po):
    return len(intersection_2(oo, po))

def get_oo_and_po_minus_co(oo, po, co):
    return len(difference_2(intersection_2(oo, po), co))

def get_po_and_co(po, co):
    return len(intersection_2(po, co))

def get_po_and_co_minus_oo(po, co, oo):
    return len(difference_2(intersection_2(po, co), oo))

def get_oo_and_co(oo, co):
    return len(intersection_2(oo, co))

def get_oo_and_co_minus_po(oo, co, po):
    return len(difference_2(intersection_2(oo, co), po))

def get_oo_and_po_and_co(oo, po, co):
    return len(intersection_3(oo, po, co))
```
